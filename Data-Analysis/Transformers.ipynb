## Add additional columns to participant CSV files

```python
# Put CSV file names for Version 1 participants in here
v1_participants = ['100_2016_Mar_29_1328.csv']

# Read CSV file for Version 1 post test with additional information that needs to get added to each participant file.
v1_df = pd.read_csv('v1_post_test_additional_info.csv')
v1_cols = v1_df.columns.values

# Go through every participant and add the columns if they do not exist.
for participant in v1_participants:
    if participant in path_list:
        df = pd.read_csv(participant)
        
        for col in v1_cols:
            if col not in df.columns:
                df[col] = v1_df[col]
        df.to_csv(participant)

# Put CSV file names for Version 2 participants in here        
v2_participants = []

# Read CSV file for Version 1 post test with additional information that needs to get added to each participant file.
v2_df = pd.read_csv('v2_post_test_additional_info.csv')
v2_cols = v2_df.columns.values

# Go through every participant and add the columns if they do not exist.
for participant in v2_participants:
    if participant in path_list:
        df = pd.read_csv(participant)
        
        for col in v2_cols:
            if col not in df.columns:
                df[col] = v2_cols[col]
        df.to_csv(participant)
```

```python
new_df = new_df.reset_index(level=0, drop=True)
```

```python
#seeing if the key response column contains a list value or an int 
list_indexes = new_df['key_resp_9.keys'].str.contains('[', regex=False, na=False)
list_series = new_df.ix[list_indexes]['key_resp_9.keys']
```

```python
#since the items were strings we changed the items to list values (ast) and got the last item in the list (-1) and that
#allowed for the the converstion to integers. 
import ast
my_list = [int(ast.literal_eval(i)[-1]) for i in list_series]
my_list
#moving back into the series (list index) 
new_series = pd.Series(my_list, index=list_series.index)
new_series.head()
```

```python
new_df.ix[list_indexes, 'key_resp_9.keys'] = new_series
new_df.head()
```
